```python
import sympy
sympy.init_printing()

u_max, u_star, rho_max, rho_star, A, B = \
    sympy.symbols(('u_max u_star rho_max rho_star A B'))
```

```python
eq1 = sympy.Eq( 0, u_max*rho_max*(1 - A*rho_max-B*rho_max**2))
eq2 = sympy.Eq( 0, u_max*(1 - 2*A*rho_star-3*B*rho_star**2))
eq3 = sympy.Eq(u_star, u_max*(1 - A*rho_star - B*rho_star**2))
```

```python
eq1
```

```python
eq2
```

```python
eq3
```

```python
eq4 = sympy.Eq(eq2.lhs - 3*eq3.lhs, eq2.rhs - 3*eq3.rhs)
eq4
```

```python
eq4.simplify()
```

```python
eq4.expand()
```

```python
rho_sol = sympy.solve(eq4, rho_star)[0]
rho_sol
```

```python
B_sol = sympy.solve(eq1,B)[0]
B_sol
```

```python
quadA = eq2.subs([(rho_star, rho_sol), (B,B_sol)])
quadA
```

```python
quadA.simplify()
```

```python
A_sol = sympy.solve(quadA, A)
```

```python
def computeF(u_max, rho, aval, bval):
    return u_max*rho*(1 - aval*rho-bval*rho**2)
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib import rcParams
rcParams['font.family'] = 'serif'
rcParams['font.size'] = 16
```

```python
def rho_green_light(nx, rho_light):
    """Computes "green light" initial condition with shock,
    and linear distribution behind
    
    Parameteres
    -----------
    nx         : int
        Number of grid points in x
    rho_light  : float
        Density of cars at stoplight
        
    Returns
    -------
    rho_initial: array of floats
        Array with initial values of density
    """
    
    rho_initial = np.arange(nx)*2./nx*rho_light  # Before stoplight
    rho_initial[(nx-1)/2:] = 0
    
    return rho_initial
```

```python
#Basic inital condition parameters
#defining grid size, time steps
nx = 81
nt = 30
dx = 4.0/(nx-1)

x = np.linspace(0,4,nx)

rho_light = 5.5
```

```python
rho_initial = rho_green_light(nx, rho_light)
```

```python
plt.plot(x, rho_initial, color='#003366', ls='-', lw=3)
plt.ylim(-0.5,11.);
```

```python
def ftbs(rho, nt, dt, dx, rho_max, u_max):
    """ Computes the solution with forward in time,
    backward in space
    
    Parameters
    ----------
    rho     : array of floats
            Density at current time-step
    nt      : int
            Number of time steps
    dt      : float
            Time-step size
    dx      : float
            Mesh spacing
    rho_max : float
            Maximum allowed car density
    u_max   : float
            Speed limit
            
    Returns
    -------
    rho_n : array of floats
            Density after nt time steps ar every point x
    """
    
    #initialize our results array with dimensions nt by nx
    rho_n = np.zeros((nt,len(rho)))
    
    # copy the initial u array into each row of our new array
    rho_n[0,:] = rho.copy()
    
    for t in range(1,nt):
        F = computeF(u_max, rho, aval, bval)
        rho_n[t,1:] = rho[1:] - dt/dx*(F[1:]-F[:-1])
        rho_n[t,0] = rho[0]
        rho_n[t,-1] = rho[-1]
        rho = rho_n[t].copy()
        
    return rho_n
```

```python
sigma = 1.
dt = sigma*dx/u_max

rho_n = ftbs(rho_initial, nt, dt, dx, rho_max, u_max)
```
