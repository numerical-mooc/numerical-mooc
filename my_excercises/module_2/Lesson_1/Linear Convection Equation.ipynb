```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib import rcParams
rcParams['font.family'] = 'serif'
rcParams['font.size'] = 16
```

```python
nx = 111
dx = 2/(nx-1)
nt = 25
dt = 0.02
c = 1       # assume wavespeed of c = 1
x = np.linspace(0,2,nx)
print(dx)
```

```python
# Initial conditions

u = np.ones(nx)
lbound = np.where(x >= 0.5)
ubound = np.where(x <= 1)

#print(lbound)
#print(ubound)
```

```python
# combine lbound and ubound array to form initial square wave

bounds = np.intersect1d(lbound, ubound)
u[bounds]=2  # setting u =2 between 0.5 and 1 as per out I.C.s
#print(u)
```

```python
plt.plot(x, u, color='#003366', ls='--', lw=3)
plt.ylim(0, 2.5);
```

Not perfectly square because there is no point where x equals 1 and 0.5?

```python
# Moving the wave in time using chosen finite-difference scheme

for n in range(1,nt):
    un = u.copy() # array to temporarily store solution of u(n+1)
    for i in range(1,nx):
        
        u[i] = un[i]-c*dt/dx*(un[i]-un[i-1])
```

```python
plt.plot(x, u, color='#003366', ls='--', lw=3)
plt.ylim(0,2.5);
```
