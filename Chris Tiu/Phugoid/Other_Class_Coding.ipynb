```python
from math import sin, cos, log, ceil
import numpy
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
Circulation = 1.0


def Lamb_Oseen_Vortex(radius, nue_time):
    answer = Circulation/(2*numpy.pi*radius)*(1-numpy.exp(-(radius**2)/4*nue_time))
    return answer
    
```

```python
R=numpy.linspace(.01, 50, 50)
NUE_TIME=1
Lamb_Oseen_Vortex(R,NUE_TIME )

plt.plot(R,Lamb_Oseen_Vortex(R,NUE_TIME ))
```
