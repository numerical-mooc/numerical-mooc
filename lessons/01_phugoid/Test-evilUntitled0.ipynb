```python
def euler_method(u,z,dt,zt=100,N=10000):
    for n in range(1,N):
        u=u + dt*numpy.array([u[1], g*(1-u[0]/zt)])
        z[n] = u[0]
```

```python
dt_values=np.array([0.1,0.05,0.01,0.005,0.001,0.0001])

# array that will contain soltion of eache grid

z_values=np.empty_like(dt_values,dtype=np.ndarray)

for i, dt in enumerate(dt_values):
    N=int(T/dt)+1   # number of time-steps
    
    ### discritiza the time using168. np.linespace()
    t=np.linspace(0.0,T,N)
    
    #initlal conditions
    u=np.array([z0,v])
    z=np.empty_like(t)
    z[0]=z0
    
    #time loop - Euler metthod
    
    #for n in range(1,N):
     #   u = u + dt*np.array([u[1],g*(1-u[0]/zt)])
      #  z[n]=u[0]
    euler_method(u,z,dt,zt,N)
    z_values[i]=z.copy()
```

```python
def get_erorr(z,dt):
    N=len(z)
    t=np.linspace(0.0,T,N)
    z_exact=v*(zt/g)**.5*numpy.sin((g/zt)**.5*t)+\
                (z0-zt)*numpy.cos((g/zt)**.5*t)+zt
    return dt*np.sum(np.abs(z-z_exact))
```

```python
error_values = np.empty_like(dt_values)

for i, dt in enumerate(dt_values):
    error_values[i] = get_erorr(z_values[i],dt)
```

```python
plt.figure(figsize=(10, 6))
plt.tick_params(axis='both', labelsize=14) #increase tick font size
plt.grid(True)                         #turn on grid lines
plt.xlabel('$\Delta t$', fontsize=16)  #x label
plt.ylabel('Error', fontsize=16)       #y label
plt.loglog(dt_values, error_values, 'ko-')  #log-log plot
plt.axis('equal')                      #make axes scale equally;
```

```python
from math import sin, cos, log,ceil
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib import rcParams
rcParams['font.family']='serif'
rcParams['font.size']=16
```

```python
# model parameters:
g = 9.8   # gravity in m s^{-2}
v_t = 30.0  # trim velocity in m s^{-1}
C_D = 1/40   # drag coefficient -- or D/L if C_L = 1
C_L = 1.0    # for conveience, use C_L = 1

### set initial conditions ### 
v0=v_t       #start at the trim velocity
theta0=0.0   # initial angle of tragectory
x0 = 0.0    #horizontal position is arbitrary
y0 = 1000.0  #initial altitude
```

```python
def f(u):
    """ Returns the right-hand-side of the phugoid system of equation
    
    Parameters
    -----------
    u: array of float
        array containing the solution at time n;
    Returns
    --------
    dudt: array of float
        array containing the RHS given u
    """
    v = u[0]
    theta = u[1]
    x = u[2]
    y = u[3]
    
    return numpy.array([-g*sin(theta)-C_D/C_L*g/(v_t**2)*(v**2),
                        -g*cos(theta)/v+g/(v_t**2)*v,
                        v*cos(theta),
                        v*sin(theta)])
```

```python
def euler_setup(u,f,dt):
    """Returns the solution at the next time-step using Euler's method.
    
    Parameters
    ----------
    u : array of float
        solution at the previous time-step.
    f : function
        function to compute the right hand-side of the system of equation.
    dt : float
        time-increment.
    
    Returns
    -------
    u_n_plus_1 : array of float
        approximate solution at the next time step.
    """
    return u + dt * f(u)
```

```python
T =100.0    #  final time
dt =0.1     # time increment
N=int(T/dt)+1  #number of tim-steps
t = numpy.linspace(0.0,T,N)  # time discretization

# initialize the array containing the solution for each time-step
u=np.empty((N,4))
u[0] =np.array([v0,theta0,x0,y0])    # fill 1st element with initial values

#time loop -Euler method
for n in range(N-1):
    
    u[n+1] = euler_setup(u[n],f,dt)
```

```python
# get teh glider's position with respect to time
x = u[:,2]
y = u[:,3]
v=u[:,0]
v
```

```python
#visualization of the path
plt.figure(figsize=(8,6))
plt.grid(True)
plt.xlabel(r'x',fontsize = 18)
plt.ylabel(r'y',fontsize = 18)
plt.title("Glider trajectory, flight time = %.2f" % T, fontsize = 18)
plt.plot(x,y,'k-',lw=2);
```
